# 2) DATA AUGMENTATION

```python
from skimage.transform import rotate
from skimage.transform import warp
from skimage.transform import SimilarityTransform

nb_datas = 1200
nb_per_class = int(nb_datas/3)
data_exp = np.zeros( (nb_datas, data.shape[1], data.shape[2]) ).astype(bool)

select_pic = np.random.randint(low = 0, high = 9, size = (nb_datas))
angles = np.random.randint(low = 0, high = 180, size = nb_datas)       #integers are enough, 1 degree is very small
trans = np.random.uniform(low = -3, high = 3, size = (nb_datas,2) )    #float as translation like 1.5 are also useful ! 
```

```python
for i in np.arange(nb_per_class):
    T = SimilarityTransform( translation = trans[i] )
    data_exp[i] = rotate(data[select_pic[i]], angles[i])
    data_exp[i] = warp( data_exp[i], inverse_map = T.inverse) 
  
for i in np.arange(nb_per_class,2*nb_per_class):
    T = SimilarityTransform( translation = trans[i] )
    data_exp[i] = rotate(data[select_pic[i]+30], angles[i])
    data_exp[i] = warp( data_exp[i], inverse_map = T.inverse) 
  
for i in np.arange(2*nb_per_class, 3*nb_per_class):
    T = SimilarityTransform( translation = trans[i] )
    data_exp[i] = rotate(data[select_pic[i]+40], angles[i])
    data_exp[i] = warp( data_exp[i], inverse_map = T.inverse) 
```

```python
# ---- Take a look of the Expanded dataset
nb = 2 #0,1 or 2
plot_pic( [ data_exp[i] for i in np.arange(nb*nb_per_class,nb*nb_per_class+30)], NB_COL = 6)
```

```python
def classify_getSymbol( pic, path_classifier='symbol_classifier_V1.pk' ):
  lda_model = pickle.load(open(path_classifier,'rb'))
  features = np.zeros((1,3))
  bend_the_knee_to_the_chosen_fourier_descriptor_feature = 5 #Heretics will burn.
  features[0,0:2] = compute_nb_objects(pic)
  features[0,2],_,_ = compute_feat_with_fourier_descriptor(pic, feat_n = bend_the_knee_to_the_chosen_fourier_descriptor_feature)
  features[0,2]*=300 #lazy wait to not normalize the data
  pred = lda_model.predict(features[:,1:3])

  if features[0,0] == 3:   # 3 points in object => Divide symbol
    return '/'
  elif features[0,0] ==2:  # 2 points in object => '=' symbol
    return '='
  elif pred[0] == 1:       # 1 point in object => use lda classifier to decide between + , - ,* symbols
    return '*'
  elif pred[0] == 2:
    return '+'
  elif pred[0] == 3:
    return '-'
  else:
    return 'STH WENT WRONG'
```
